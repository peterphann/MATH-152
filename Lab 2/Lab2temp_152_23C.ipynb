### Question 1

#### 1a

```python
#start code here
# Rotate around y-axis = y integral
y = sp.symbols('y')
f_cone = y / 4
f_above = sp.sqrt(1 - (y-4)**2)
volume_cone = sp.integrate(sp.pi * f_cone**2, (y, 0, 4))
volume_above = sp.integrate(sp.pi * f_above**2, (y, 4, 5))
total_volume = volume_cone.evalf() + volume_above.evalf()
print('The volume of the cone by rotating about the y-axis is', total_volume)
```

#### 1b

```python
#start code here
r = sp.symbols('r')
u_cone = (r / 4) * y
u_volume = sp.integrate(sp.pi * u_cone**2, (y, 0, 4))
solution = sp.solve(u_volume - 14.5, r)
print('The radius of the cone needed for a cone of volume 14.5 cubic inches is', solution[1], 'in')
```

```python
#start code here
x = sp.symbols('x', real=True)
f_x = 2 * (x ** 2)
g_x = abs(x ** 2 - 9)

plot(f_x, g_x, (x, -5, 5))
```

#### 2b

```python
#start code here
solutions = sp.solve(f_x - g_x, x)
a = solutions[0]
b = solutions[1]
print('The points of intersection of the two curves are', a, 'and', b)
```

#### 2c

```python
#start code here
area = sp.integrate(g_x - f_x, (x, a, b))
print('The area of the region enclosed by the two curves is', area.evalf())
```
