```python
import numpy as np
import sympy as sp
from sympy.plotting import (plot, plot_parametric)
```

### Question 1

#### 1a

```python
# start your code here
x = sp.symbols('x')
f = 1 / (sp.sqrt(2 * sp.pi)) * sp.exp((-1 / 2) * (x ** 2))
plot(f, (x, -3, 3))
```

#### 1b Left Endpoint Approximation

```python
# start your code here
a = -1
b = 1
n = 500
dx = (b - a) / n
xi = np.arange(a, b, dx)
yi = [f.subs(x, i) for i in xi]

left_sum = sum(yi) * dx
print(left_sum.evalf())
```

### Question 2 Right Endpoint Approximation

```python
# start your code here
xi = np.arange(a + dx, b + dx, dx)
yi = [f.subs(x, i) for i in xi]

right_sum = sum(yi) * dx
print(right_sum.evalf())
```

### Question 3 Midpoint Approximation

#### 3a

```python
# start your code here
# start your code here
xi = np.arange(a + (dx / 2), b + (dx / 2), dx)
yi = [f.subs(x, i) for i in xi]

mid_sum = sum(yi) * dx
print(mid_sum.evalf())
```

#### 3b

```python
# start your code here
average = (left_sum.evalf() + right_sum.evalf()) / 2
print(f'The average of the left and right endpoint approximations is {average}.')
print('This is not equal to the midpoint approximation.')
```

### Question 4 Trapezoid Approximation

#### 4a

```python
# start your code here
from scipy.integrate import trapz

xi = np.arange(a, b + dx, dx)
yi = [f.subs(x, i) for i in xi]
trapz_sum = trapz(yi, xi)
print(trapz_sum.evalf())
```

#### 4b

```python
# Write your answer as either a comment or a print statement.
print('Yes, the trapezoid approximation is equal to the average of the left and right endpoint approximations.')
```

### Question 5 Simpson's Rule

```python
# start your code here
from scipy.integrate import simps

xi = np.arange(a, b + dx, dx)
yi = [f.subs(x, i) for i in xi]
simp_sum = simps(yi, xi)
print(simp_sum.evalf())
```

### Question 6 Errors

```python
# start your code here
actual = 0.682689492137086

error1 = abs(left_sum.evalf() - actual)
error2 = abs(right_sum.evalf() - actual)
error3 = abs(mid_sum.evalf() - actual)
error4 = abs(trapz_sum.evalf() - actual)
error5 = abs(simp_sum.evalf() - actual)

print(f'The error for the left endpoint approximation is {error1}.')
print(f'The error for the right endpoint approximation is {error2}.')
print(f'The error for the midpoint approximation is {error3}.')
print(f'The error for the trapezoidal approximation is {error4}.')
print(f'The error for the Simpson\'s approximation is {error5}.')
```
